```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
class StringConverter(dict):
    def __contains__(self, item):
        return True

    def __getitem__(self, item):
        return str

    def get(self, default=None):
        return str
```

```python
bank_accounts = pd.read_csv("fraud_shopee\\bank_accounts.csv", converters=StringConverter())
credit_card = pd.read_csv("fraud_shopee\\credit_cards.csv", converters=StringConverter())
devices = pd.read_csv("fraud_shopee\\devices.csv", converters=StringConverter())
orders = pd.read_csv("fraud_shopee\\orders.csv", converters=StringConverter())
```

B1: Lấy ra các userid từ bảng orders tạo thành df userids

```python
userids = pd.DataFrame(np.concatenate([orders['buyer_userid'], orders['seller_userid']]), columns=['userid'])
userids = pd.DataFrame(userids['userid'].unique(), columns=['userid'])
userids
```

B2: Join bảng userid với devices, credit_card, bank_account

```python
userid_df = userids.merge(devices, how='left', on='userid')
userid_df = userid_df.merge(credit_card, how='left', on='userid')
userid_df = userid_df.merge(bank_accounts, how='left', on='userid')
userid_df.head()
```

B3: So khớp để tìm

```python
def isFraud(buyerid, sellerid):
    userA = master_userid_df[master_userid_df['userid'] == str(buyerid)]
    userB = master_userid_df[master_userid_df['userid'] == str(sellerid)]

    # devices
    devicesA = userA['device'].values[0]
    devicesB = userB['device'].values[0]

    # credit_card
    credit_cardA = userA['credit_card'].values[0]
    credit_cardB = userB['credit_card'].values[0]

    # bank_account
    bank_accountA = userA['bank_account'].values[0]
    bank_accountB = userB['bank_account'].values[0]

    if not ({np.nan}.issubset(devicesA) or {np.nan}.issubset(devicesB)):
        if any([x in devicesA for x in devicesB]):
            return True

    if not ({np.nan}.issubset(credit_cardA) or {np.nan}.issubset(credit_cardB)):
        if any([x in credit_cardA for x in credit_cardB]):
            return True

    if not ({np.nan}.issubset(bank_accountA) or {np.nan}.issubset(bank_accountB)):
        if any([x in bank_accountA for x in bank_accountB]):
            return True

    return False
```

B6: Triển khai Code trường hợp Gián tiếp.

Chúng ta sẽ cần các master_bank, master_credit, master_device và master_user

```python
master_bank = bank_accounts.groupby('bank_account')
master_bank = master_bank.agg(lambda x: set(x))
master_bank = master_bank.reset_index()
```

```python
master_bank
```

```python
master_credit = credit_card.groupby('credit_card')
master_credit = master_credit.agg(lambda x: set(x))
master_credit = master_credit.reset_index()
```

```python
master_credit
```

```python
master_devices = devices.groupby('device')
master_devices = master_devices.agg(lambda x: set(x))
master_devices = master_devices.reset_index()
```

```python
master_devices
```

```python
def isFraud_v2(buyer_is, seller_id, depth_search = 3):
    userA = buyer_is
    userB = seller_id

    user_relayed = set()
    user_relayed.update({userA})

    for i in range(depth_search):
        ### devices
        devices_relayed = set()
        for user in user_relayed:
            temp = master_userid_df[master_userid_df['userid'] == user]
            if temp.empty:
                continue
            else:
                devices_relayed.update(temp['device'].values[0])
        # print(devices_relayed)

        user_relayed = set()
        for device in devices_relayed:
            temp = master_devices[master_devices['device'] == device]
            if temp.empty:
                continue
            else:
                user_relayed.update(temp['userid'].values[0])
        # print(user_relayed)

        if userB in user_relayed:
            return True

        ### bank_account
        bank_relayed = set()
        for user in user_relayed:
            temp = master_userid_df[master_userid_df['userid'] == user]
            if temp.empty:
                continue
            else:
                bank_relayed.update(temp['bank_account'].values[0])
        # print(bank_relayed)

        user_relayed = set()
        for bank in bank_relayed:
            temp = master_bank[master_bank['bank_account'] == bank]
            if temp.empty:
                continue
            else:
                user_relayed.update(temp['userid'].values[0])
        # print(user_relayed)

        if userB in user_relayed:
            return True

        ### credit_account
        credit_relayed = set()
        for user in user_relayed:
            temp = master_userid_df[master_userid_df['userid'] == user]
            if temp.empty:
                continue
            else:
                credit_relayed.update(temp['credit_card'].values[0])
        # print(credit_relayed)

        user_relayed = set()
        for credit in credit_relayed:
            temp = master_credit[master_credit['credit_card'] == credit]
            if temp.empty:
                continue
            else:
                user_relayed.update(temp['userid'].values[0])
        # print(user_relayed)

        if userB in user_relayed:
            return True

    return False
```

```python
for i in range(len(orders[:10])):
    if isFraud_v2(orders.iloc[i].values[1], orders.iloc[i].values[2], depth_search=10):
        print(i)
```

```python
def cal_threading1(orders):
    for i in range(len(orders)):
        if isFraud_v2(orders.iloc[i].values[1], orders.iloc[i].values[2], depth_search=10):
            print('1', 'Fraud: ', orders.iloc[i].values[0])
        else:
            print('1', 'Not - Fraud: ', orders.iloc[i].values[0])

def cal_threading2(orders):
    for i in range(len(orders)):
        if isFraud_v2(orders.iloc[i].values[1], orders.iloc[i].values[2], depth_search=10):
            print('2', 'Fraud: ', orders.iloc[i].values[0])
        else:
            print('2', 'Not - Fraud: ', orders.iloc[i].values[0])
```

```python
from threading import Thread
import threading
import time

try:
    t1 = threading.Thread(target=cal_threading1, args=(orders[:6],))
    t2 = threading.Thread(target=cal_threading2, args=(orders[6:10],))
    t = time.time()
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    print('Time: ', time.time() - t)
except:
    print('Error')
```
