## Lấy ra toàn bộ userid, device, bank_account,credit_card trong 3 bảng device, bank_account, credit_card để tạo các node

### Userid

```python
userid = pd.concat((bank_account_df['userid'], devices_df['userid'], credit_card_df['userid']), ignore_index=True)
userid
```

```python
userid = pd.DataFrame(userid.drop_duplicates().reset_index(drop=True))
userid
```

```python
# Sort to define the order of nodes
sort_userid = userid.sort_values(by='userid').set_index('userid')

# Map IDs to start from 0
sort_userid = sort_userid.reset_index(drop=False)
user_id_mapping = sort_userid["userid"]
```

```python
user_id_mapping
```

### Device

Lấy ra device từ bảng devices

```python
devices = devices_df['device']
devices = pd.DataFrame(devices.drop_duplicates().reset_index(drop=True))
devices
```

```python
devices_id_mapping = devices['device']
devices_id_mapping
```

### Bank_accounts

```python
bank_account = bank_account_df['bank_account']
bank_account = pd.DataFrame(bank_account.drop_duplicates().reset_index(drop=True))
bank_account
```

```python
bank_account_id_mapping = bank_account['bank_account']
bank_account_id_mapping
```

### Credit_card

```python
credit_card = credit_card_df['credit_card']
credit_card = pd.DataFrame(credit_card.drop_duplicates().reset_index(drop=True))
credit_card
```

```python
credit_card_id_mapping = credit_card['credit_card']
credit_card_id_mapping
```

## Xây dựng Mapping bằng chỉ số index của các bảng *_df

### Mapping of device

```python
devices_userid_mapping = devices_df.copy(deep=True)
devices_userid_mapping
```

```python
device_map = devices_id_mapping.reset_index().set_index('device').to_dict()
devices_userid_mapping['device'] = devices_userid_mapping['device'].map(device_map['index']).astype(int)
# devices_userid_mapping
```

```python
userid_map = user_id_mapping.reset_index().set_index('userid').to_dict()
devices_userid_mapping['userid'] = devices_userid_mapping['userid'].map(userid_map['index']).astype(int)
devices_userid_mapping
```

```python
edge_index_device = devices_userid_mapping[['userid', 'device']].values.transpose()
edge_index_device
```

### Mapping of Bank_account

```python
bank_account_mapping = bank_account_df.copy(deep=True)
bank_account_mapping
```

```python
bank_accounts_map = bank_account_id_mapping.reset_index().set_index('bank_account').to_dict()
bank_account_mapping['bank_account'] = bank_account_mapping['bank_account'].map(bank_accounts_map['index']).astype(int)
# bank_account_mapping
```

```python
userid_map = user_id_mapping.reset_index().set_index('userid').to_dict()
bank_account_mapping['userid'] = bank_account_mapping['userid'].map(userid_map['index']).astype(int)
bank_account_mapping
```

```python
edge_index_bank = bank_account_mapping[['userid', 'bank_account']].values.transpose()
edge_index_bank
```

### Mapping of Credit_card

```python
credit_card_mapping = credit_card_df.copy(deep=True)
credit_card_mapping
```

```python
credit_card_map = credit_card_id_mapping.reset_index().set_index('credit_card').to_dict()
credit_card_mapping['credit_card'] = credit_card_mapping['credit_card'].map(credit_card_map['index']).astype(int)
# credit_card_mapping
```

```python
userid_map = user_id_mapping.reset_index().set_index('userid').to_dict()
credit_card_mapping['userid'] = credit_card_mapping['userid'].map(userid_map['index']).astype(int)
credit_card_mapping
```

```python
edge_index_credit = credit_card_mapping[['userid', 'credit_card']].values.transpose()
edge_index_credit
```

## Build dataset

```python
from torch_geometric.data import HeteroData
data = HeteroData()

data['user'].x = userid
data['device'].x = devices
data['bank_account'].x = bank_account
data['credit_card'].x = credit_card

data['user', 'rating', 'device'].edge_index = edge_index_device
data['user', 'rating', 'bank_account'].edge_index = edge_index_bank
data['user', 'rating', 'credit_card'].edge_index = edge_index_credit
```
